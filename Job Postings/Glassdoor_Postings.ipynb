```python
#import libraries

import numpy as np
import pandas as pd
import re



pd.set_option('display.max_rows', 99)
```

```python
#import file

df = pd.read_csv('Uncleaned_DS_jobs.csv')

df

```

```python
df.info()
```

```python
#extract only figures from 'Salary Estimate'

pattern = r'\$(\d+)K.\$(\d+)K'

df['Salary Estimate'].str.extract(pattern)


```

```python
#create new columns for the 2 columns and convert to interger

df[['Low', 'High']] = df['Salary Estimate'].str.extract(pattern)



df[['Low', 'High']] = df[['Low', 'High']].map(int)
```

```python
#check the data type
df[['Low', 'High']].dtypes
```

```python
#merge converted columns into a single column

df['Salary'] = df['Low'].astype(str) + "-" + df['High'].astype(str)

df
```

```python
#Inspect Job description

df.loc[1,'Job Description']
```

```python
df
```

```python
#Inspect rating and company name

df[['Rating', 'Company Name']].head(20)
```

```python
#Remove the rating from company name

df['Company'] = df.apply(lambda a: a['Company Name'] if a['Rating']<0 else a['Company Name'][:-4],axis=1)
df.head(20)
```

```python
df
```

```python
df['Location']
```

```python
#get the total for each location

df['Location'].value_counts()
```

```python
#extract the state from location into a new column

state_pattern = r',\s([A-Z]{2})'
df['State']= df['Location'].str.extract(state_pattern, expand=False)
df

```

```python
#Skills in Demand

df['Python'] = df['Job Description'].apply(lambda b: 1 if 'python' in b else 0)
df['R'] = df['Job Description'].apply(lambda b: 1 if 'r' in b else 0)
df['Excel'] = df['Job Description'].apply(lambda b: 1 if 'excel' in b else 0)
df['AWS'] = df['Job Description'].apply(lambda b: 1 if 'aws' in b else 0)
df['SQL'] =df['Job Description'].apply(lambda b: 1 if 'sql' in b else 0)
df['Statistics'] =df['Job Description'].apply(lambda b: 1 if 'statistics' in b else 0)
df['Machine Learning'] = df['Job Description'].apply(lambda b: 1 if 'machine learning' in b else 0)
```

```python
df
```

```python
df['Avg_Salary'] = (df['Low'] + df['High']) / 2
df
```

```python
df = df.rename(columns={'Low': 'Lowest_Salary', 'High': 'Highest_Salary'})
```

```python
#select only needed data

data_columns = ['Job Title', 'Job Description', 'Rating', 'Headquarters', 'Size', 'Sector', 'Revenue', 'Salary','Lowest_Salary', 'Highest_Salary','Avg_Salary', 'Company', 'State', 'Python', 'R', 'Excel', 'AWS', 'SQL', 'Statistics', 'Machine Learning']

df1 = df[data_columns].copy()

df1.head()
```

```python
#save cleaned data to csv

df1.to_csv('Cleaned_JP.csv')
```
