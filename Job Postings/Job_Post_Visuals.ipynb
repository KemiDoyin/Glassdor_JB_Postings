```python
#Lets look at Job Openings that need Data Scientist and their Average Salaries

DataSci = df[df['Job Title'].str.contains('Data Scientist|Data Science', case=False, regex=True)]

count_DS = DataSci.groupby(['Job Title', 'Avg_Salary']).size().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')

table = ax.table(cellText=count_DS.reset_index()[['Job Title', 'Avg_Salary']].values,
                            colLabels=['Job Title', 'Avg_Salary'],
                            cellLoc='center',
                            loc='center')

table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  

plt.show()
```

```python
#Look into the job sectors
df['Sector'].unique()
```

```python
#Popular Organisation Sector for Job Openings

count_sector = df['Sector'].value_counts()[:9]

count_sector.plot(kind='bar')

plt.ylabel('Count')
plt.title('Count of Available Job Sectors')

plt.show()
```

Looking at the last two visuals, we can see that Jobs in Consumer Services pay better that any other but there are not much openings for it compared to IT , Business Services, HealthCare....




Let's take a look at the skills that are mostly required for Data Science/Engineering Jobs as stated in Glassdoor

```python
skills_needed = df[df.columns[df.eq(1).any()]]



count_skills = skills_needed.sum()



plt.figure(figsize=(8, 8))

plt.pie(count_skills, autopct='%1.1f%%', startangle=130)

plt.title('Distribution of Skills')

plt.legend(labels=count_skills.index, loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```

From the pie-chart above, most job postings require *R* as an important skill to have.  *Machine Learning*, *Statistics*, and *Excel* are also quite popular skills that these job posting state.

*SQL*, *AWS*, *Python* are nice to have.
